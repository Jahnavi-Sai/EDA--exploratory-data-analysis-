# EDA - Exploaratory data analysis 

Exploratory Data Analysis (EDA) is the process of analyzing and summarizing datasets to uncover patterns, trends, relationships, and anomalies. It uses visualization and basic statistics to help understand the structure of the data before applying more advanced techniques or building models.

EDA is the foundation of machine learning and it is to understand the various aspects of the data.



# Main Objectives of EDA:

``1.Understand the structure and quality of the data``
(Check data types, missing values, duplicates, and basic stats)

``2.Discover patterns, relationships, and trends``
(Use visualizations and summary stats to explore variable interactions)

``3.Identify outliers and anomalies``
(Spot unusual data points that may impact analysis or models)

# Why Exploratory Data Analysis is Important?

Exploratory Data Analysis (EDA) is important because it helps you understand your data before making any assumptions or building predictive models. Here’s why EDA is essential:

``1. Identifies Data Quality Issues:`` EDA helps spot missing values, outliers, duplicate records, and inconsistent formats that could distort analysis.

``2. Reveals Patterns and Relationships:`` It uncovers trends, correlations, and distributions that provide insight into how variables interact.

``3. Guides Feature Selection:`` EDA helps in choosing the most relevant features for modeling by analyzing variable importance and redundancy.

``4. Informs Model Selection:`` By understanding data behavior (e.g., linear vs. nonlinear trends), EDA aids in selecting appropriate algorithms.

``5. Supports Hypothesis Testing:`` It allows you to form and validate assumptions with visual evidence and summary statistics.

``6. Improves Communication:`` EDA visuals (e.g., histograms, box plots, scatter plots) make it easier to explain data-driven decisions to stakeholders.

# Architecture to bulid predictue model
 
``1.Inputs (Raw Data Sources):`` 

This is the first stage where we collect raw data from different sources. These sources can vary depending on the problem you're solving.

Examples of Raw Data Sources:

• CSV Files: Common export from Excel or other systems (sales.csv, customers.csv)

• Databases: SQL (MySQL, PostgreSQL) or NoSQL (MongoDB) systems with structured or semi-structured data

• APIs / Web Scraping: Fetch live data from websites (e.g., weather, stock prices, product reviews)

• User Inputs / IoT / Logs: Data from sensors, devices, or web server logs

``2. Collated Dataset``

In this step, we integrate all the collected data into one central dataset.

• Key Actions:

1. Merge datasets based on common keys (e.g., Location ID, House ID)

2. Join different features (e.g., append crime stats to house locations)

3. Append new records if multiple sources give similar structures

• Example:

1. Merge house sales with neighborhood data (crime, schools)

2. Join with weather info if relevant

``3. Cleansing + Transformation``

Before training a model, the data must be cleaned and transformed to ensure consistency and usefulness.

• Key Tasks:

1. Handle Missing Values: Fill with mean, median, or drop rows

2. Remove Duplicates: Avoid skewing the data

3. Correct Formats: Ensure numbers are numbers, dates are parsed, etc.

4. Encode Categorical Variables: Convert text labels to numbers (One-Hot Encoding, Label Encoding)

5. Scale Features: Use StandardScaler or MinMaxScaler for numeric consistency

6. Feature Engineering: Create new meaningful variables (e.g., price per sqft)

• Example:

1. Fill missing School Rating with the average

2. Convert Location to binary columns (Zone A → [1,0,0])

3. Scale Size to a 0-1 range

``4. Final Dataset → Build Predictive Model``

Now that the dataset is clean and structured, we can train a predictive model using machine learning algorithms.

• Key Steps:

1. Split the Dataset: Typically into train and test sets (e.g., 80/20)

2. Choose an Algorithm: Linear Regression (for price predictions), Decision Trees / Random Forest, XGBoost / LightGBM (for performance and accuracy)

3. Train the Model: Use training data to fit the model

4. Evaluate the Model: Use metrics like MAE, MSE, RMSE, R², Tune hyperparameters using GridSearch or cross-validation

5. Deploy the Model: Use Flask/FastAPI, or deploy to cloud platforms (AWS, Azure, GCP)

• Example:

1. Train a regression model to predict house price based on features

2.  Evaluate with RMSE to check error

3.  Deploy the model into a web app for users to input house info and get price estimates

# 1. Variable Identification

Variable identification is one of the first and most crucial steps in Exploratory Data Analysis (EDA). It involves classifying each column (feature) in your dataset based on:

The role it plays in the dataset (input/output)

The type of data it contains (numeric, categorical, etc.)

``Types of Variables (Based on Role)``

| Type                          | Description                                   | Example                   |
| ----------------------------- | --------------------------------------------- | ------------------------- |
| **Independent (Input)**       | Variables used to predict the target variable | Age, Gender, Salary       |
| **Dependent (Output/Target)** | Variable that you want to predict             | Purchase, Price, Surviva
 |


``Types of Variables (Based on Data Type)``

» Numerical Variables
Continuous: Can take any value within a range
E.g.: height, weight, income, temperature

» Discrete: Integer countable values
E.g.: number of children, product ratings

» Categorical Variables
Nominal: Categories with no order
E.g.: gender, color, city

» Ordinal: Categories with a natural order
E.g.: low/medium/high, education level

» Date/Time Variables
Used for time-series analysis or time-based grouping
E.g.: order_date, join_date

» Boolean Variables
Binary: True/False, 0/1
E.g.: is_subscribed, has_defaulted

# 3. Bivariate Analysis:

Bivariate Analysis focuses on identifying relationship between two variables to find connections, correlations and dependencies. It helps to understand how two variables interact with each other. 

``Purpose of Bivariate Analysis:``

‣ To examine the relationship between two variables

‣ To detect patterns, trends, or associations

‣ To support hypothesis testing

‣ To prepare data for model building

``Types of Variable Combinations:``

``1. Numerical vs Numerical``

Objective: Find correlation or linear relationship

Example: Income vs Age

Techniques: Correlation coefficient, regression

Visualization: Scatter plot

``2. Numerical vs Categorical``

Objective: Compare group-wise means or medians

Example: Salary vs Gender

Techniques: Grouped summary stats, ANOVA

Visualization: Box plot, Violin plot

``3. Categorical vs Categorical``

Objective: Check association between categories

Example: Gender vs Marital Status

Techniques: Cross-tabulation, Chi-square test

Visualization: Bar plot, Stacked bar chart

``➡ Correlation:`` Measures the strength and direction of a linear relationship between two numerical variables.

| **Type**               | **Description**                              |
| ---------------------- | -------------------------------------------- |
| **Positive (+1)**      | Both variables increase or decrease together |
| **Negative (−1)**      | One variable increases, the other decreases  |
| **No Correlation (0)** | No consistent relationship between variables |

# 4. Outlier Detection:

Outliers are data points that differ significantly from other observations in the dataset. They can arise due to variability in the data, measurement errors, or experimental errors. Detecting outliers is important because they can distort statistical analyses and models.

An outlier is a value in your data that looks very different from the rest — it’s either much bigger or much smaller than most other values.

``How to Find Outliers?``

1. Look at a Box Plot

2. Use Simple Rules (IQR method)

3. Check Z-Score

``Quick Example``: If most people in a class scored between 40 and 80 on a test, but one person scored 0 or 100, that 0 or 100 might be an outlier.



# 5. Missing Value treatement:

Missing values happen when some data points are not recorded or lost. For example, a survey respondent might skip a question, or a sensor might fail to record a measurement.

``Why Treat Missing Values?``

‣ Most analysis and machine learning models can’t handle missing data well.

‣ Missing values can lead to biased or wrong results if ignored.

``1. Mean (Average)``

• What it is: Add all the numbers and then divide by how many numbers there are.

• Use it when: Data is fairly balanced without extreme values.

Formula:


``Example for mean``

Consider these numbers: 2, 4, 4, 7, 9

Mean: (2 + 4 + 4 + 7 + 9) ÷ 5 = 26 ÷ 5 = 5.2

``2. Median (Middle Value)`` 

• What it is: The middle number when the data is sorted from smallest to largest.

• Use it when: Data is skewed or has outliers (extremely high or low values).

How to find:

◦ Sort the numbers.

◦ If odd number of values, median is the middle one.

◦ If even number of values, median is the average of the two middle values.

``Example for median``

Consider these numbers: 2, 4, 4, 7, 9
​
Median: Sorted list is the same: 2, 4, 4, 7, 9

Middle number (3rd) = 4

``3.Mode (Most Frequent Value)``

• What it is: The value that appears most often.

• Use it when: Working with categorical data or to find the most common number.

``Example for mode``

Consider these numbers: 2, 4, 4, 7, 9

Mode: The number that appears most is 4 (it appears twice)

# 7. Variable creation:

``Purpose of Variable Creation``

◦ Improve model accuracy and performance.

◦ Add domain knowledge into the data.

◦ Reveal hidden patterns or relationships.
